# Visualize some protein ligand pairs to make sure the docking worked correctly



```python
from rdkit import Chem

import py3Dmol

import pandas as pd
```

```python
df = pd.read_csv('./41586_2019_917_MOESM4_ESM.csv').iloc[:-4]
inhibition = df['Inhibition (%) at 10uM'].astype(float).values
smiles = df['SMILES']
zincnames = df['ZINC ID']
```

```python
mols = []
with Chem.SDMolSupplier('docked.sdf') as sup:
    for m in sup:
        if m is not None:
            mols.append(m)
names = [m.GetProp('_Name') for m in mols]
scores = [m.GetProp('minimizedAffinity') for m in mols]
mol_df = pd.DataFrame({'scores':scores, 'names':names, 'mols':mols})
#best_conf = mol_df.sort_values('scores').drop_duplicates('names')
```

```python
view = py3Dmol.view()
view.addModel(open('./protein_5WIU.pdb').read(), 'pdb')
view.addStyle({'model':0}, {'cartoon':{'colorscheme':'orangeCarbon'}})
view.addStyle({'model':0}, {'stick':{'colorscheme':'whiteCarbon', 'radius':0.1}})



dockmols = mol_df[mol_df['names']=='ZINC000480408888']['mols']
c = 1
for mol in dockmols.iloc[:2]:
    view.addModel(Chem.MolToMolBlock(Chem.AddHs(mol,addCoords=True)), 'sdf')
    view.addStyle({'model':c}, {'stick':{'colorscheme':'blueCarbon', 'radius':0.2}})
    c+=1

view.zoomTo({'model':1})

```
